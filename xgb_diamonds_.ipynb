```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# %%
diamonds = sns.load_dataset('diamonds')

# %%
diamonds.head()

# %%
diamonds.shape

# %%
diamonds .describe()

# %%
diamonds.describe(exclude=np.number)

# %%
from sklearn.model_selection import train_test_split

# %%
# extract feature and target arrays
# Keep x as a DataFrame for now to use pandas methods
x_df, y = diamonds.drop('price', axis=1), diamonds['price'].values

# extract text features
# Use select_dtypes on the DataFrame x_df
cats = x_df.select_dtypes(exclude=np.number).columns.tolist()
# No need to convert to 'category' dtype if we are one-hot encoding
# for col in cats:
#     x_df[col] = x_df[col].astype('category')

# Perform one-hot encoding on the categorical columns
x_df_encoded = pd.get_dummies(x_df, columns=cats, drop_first=True) # drop_first=True to avoid multicollinearity

# Now convert the one-hot encoded DataFrame to a numpy array
x = x_df_encoded.values

# %%
display(x)

# %%
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# %%
import xgboost as xgb

# Since we've one-hot encoded, enable_categorical is no longer needed
dtrain_reg = xgb.DMatrix(x_train, label=y_train)
dtest_reg = xgb.DMatrix(x_test, label=y_test)
```

```python
import xgboost as xgb

# Since we've one-hot encoded, enable_categorical is no longer needed
dtrain_reg = xgb.DMatrix(x_train, label=y_train)
dtest_reg = xgb.DMatrix(x_test, label=y_test)
```

```python
import numpy as np

# Define placeholder variables for actual and predicted values
# You will replace these with actual results from your model
actual = y_test # Assuming y_test contains the actual target values
predicted = np.random.rand(len(y_test)) * y_test.max() # Replace with your model's predictions

mse = np.mean((actual - predicted)**2)
rmse = np.sqrt(mse)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
params ={"objective":"reg:squarederror","tree_method":"hist"} # Changed tree_method to 'hist' for CPU training
n = 100
model = xgb.train(
    params=params,
    dtrain=dtrain_reg,
    num_boost_round=n,
    )
```

```python
from sklearn.metrics import mean_squared_error
preds =model.predict(dtest_reg)
rmse = np.sqrt(mean_squared_error(y_test,preds))
print(f"RMSE: {rmse}")
```
